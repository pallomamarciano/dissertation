```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
keyword = 'kit gay'
maxTweets = 1000000
tweets = []
tdf = None
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(keyword + ' since:2018-07-01 lang:pt').get_items()) :
        if i > maxTweets :
            break
        username = tweet.username
        text = tweet.content
        date = tweet.date
        pubdate = tweet.date
        permalink = tweet.url
        tweets.append({
            "permalink":permalink,
            "pubdate":pubdate,
            "text":text,
            "username":username,
            "date":date
        })
```

```python
tweets
```

```python
df_tweets = pd.DataFrame.from_dict(tweets, orient='columns')
```

```python
df_tweets.shape
```

```python
df_tweets['date'].min()
```

```python
df_tweets.to_csv('h_1_20180701_0001.csv')
```
